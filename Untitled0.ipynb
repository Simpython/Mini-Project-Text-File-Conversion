```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "File Size (MB)": [200, 400, 600, 800, 1000],
    "C": [12, 20, 34, 45, 55],
    "C++": [15, 25, 36, 50, 60],
    "Java": [18, 30, 40, 55, 70],
    "R": [20, 35, 45, 60, 80],
    "Python": [25, 40, 53, 75, 100]
}

# Create table
df = pd.DataFrame(data)

# Display table
print(df)

# Plot graph
plt.plot(df["File Size (MB)"], df["C"], marker='o', label="C")
plt.plot(df["File Size (MB)"], df["C++"], marker='s', label="C++")
plt.plot(df["File Size (MB)"], df["Java"], marker='^', label="Java")
plt.plot(df["File Size (MB)"], df["R"], marker='x', label="R")
plt.plot(df["File Size (MB)"], df["Python"], marker='*', label="Python")

plt.title("Performance Comparison")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (sec)")
plt.legend()
plt.grid(True)

plt.show()
```
